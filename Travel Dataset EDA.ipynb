```python
df["TypeofContact"] = df["TypeofContact"].fillna(df["TypeofContact"].mode()[0])
```

```python
df["NumberOfTrips"] = df["NumberOfTrips"].fillna(0)
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["NumberOfChildrenVisiting"] = df["NumberOfChildrenVisiting"].fillna(df["NumberOfChildrenVisiting"].mean()).apply(np.floor)
df["MonthlyIncome"] = df["MonthlyIncome"].fillna(df["MonthlyIncome"].mean())
df["PreferredPropertyStar"] = df["PreferredPropertyStar"].fillna(df["PreferredPropertyStar"].mean()).apply(np.floor)
df["DurationOfPitch"] = df["DurationOfPitch"].fillna(df["DurationOfPitch"].median())
df["NumberOfFollowups"] = df["NumberOfFollowups"].fillna(df["NumberOfFollowups"].median())
```

# A thread on the Replacement of the missing values
df["NumberOfTrips"] = df["NumberOfTrips"].fillna(0)
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["NumberOfChildrenVisiting"] = df["NumberOfChildrenVisiting"].fillna(df["NumberOfChildrenVisiting"].mean()).apply(np.floor)
df["MonthlyIncome"] = df["MonthlyIncome"].fillna(df["MonthlyIncome"].mean())
df["PreferredPropertyStar"] = df["PreferredPropertyStar"].fillna(df["PreferredPropertyStar"].mean()).apply(np.floor)
df["DurationOfPitch"] = df["DurationOfPitch"].fillna(df["DurationOfPitch"].median())
df["NumberOfFollowups"] = df["NumberOfFollowups"].fillna(df["NumberOfFollowups"].median())
1. Missing Values in Number of trips has been filled with 0 as that is the inference that the lead may have not taken any trips till date
2. Missing  values in Age are replaced by the mean of the data as its safe to assume that the average age of the travellers can be considered for the analysis
3. Missing Values in Number of childern visiting are replaced by the mean of the data as its safe to assume that the average of the Number of childern visiting of the travellers can be considered for the analysis
4. Missing Values in Monthly Income are replaced by the mean of the data as its safe to assume that the average of the Monthly Income of the travellers can be considered for the analysis
5. Missing Values in Preferred Property Star are replaced by the floor of the mean of the data as its safe to assume that the average of the Preferred Property Star of the travellers can be considered for the analysis. This can also be done using the median which is more appropriate to use in this case scenario. But as the floor function is used,it'll work just as the median function on the dataset
6. Missing Values in Duration of Pitch are replaced by the median of the data as its safe to assume that the median value of the Duration of Pitch of the travellers can be considered for the analysis
7. Missing Values in Number of followups are replaced by the median of the data as its safe to assume that the median value of the Number of followups of the travellers can be considered for the analysis
